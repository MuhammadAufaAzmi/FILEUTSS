```
import pandas as pd
from sklearn.model_selection import train_test_split

# Load the datasets, ensuring proper variable names.
dataset1 = pd.read_csv("DDoS ICMP Flood.csv")  # Assuming this is the first dataset
dataset2 = pd.read_csv("DDoS UDP Flood.csv")  # Assuming this is the second dataset
dataset3 = pd.read_csv("DoS ICMP Flood.csv")  # Assuming this is the third dataset

hasilgabung = pd.concat([dataset1, dataset2, dataset3], ignore_index=True)

# Assuming 'Label' is the name of the target variable column in your dataset
y = hasilgabung['Label']  # Replace 'Label' with the actual column name

x = hasilgabung.iloc[:, 7:76]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
```

```
alya = DecisionTreeClassifier(criterion='entropy', splitter = 'random')
alya.fit(x_train,y_train)
```

```
y_pred = alya.predict(x_test)
```

```
y_pred
```

```
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test,y_pred)
```

```
accuracy
```

```
import matplotlib.pyplot as plt
import numpy as np
```

```
fig = plt.figure(figsize = (10, 7))
tree.plot_tree(alya, feature_names = x.columns.values, class_names = np.array([ 'Benign Traffic','DDos ICMP Flood','DDoS UDP Flood']), filled = True)
plt.show()
```

```
import seaborn as lol
from sklearn import metrics
label = np.array([ 'MQTT Malformed','Recon OS Scan','Recon Ping Sweep'])
```

```
import matplotlib.pyplot as plt
conf_matrix = metrics.confusion_matrix(y_test, y_pred)
plt.figure(figsize=(10, 10))
lol.heatmap(conf_matrix, annot=True, xticklabels=label, yticklabels=label)
plt.xlabel('Prediksi')
plt.ylabel('Fakta')
plt.show()
```
